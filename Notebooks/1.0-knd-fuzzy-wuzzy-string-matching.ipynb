# String Matching Company Names Between Documents

```python
import pandas as pd  
import matplotlib.pyplot as plt 

from fuzzywuzzy import process, fuzz 
```

```python
#import raw data
masterTickers = pd.read_csv("Data/Raw/masterTickers.csv")
checkTheseTickers = pd.read_csv("Data/Raw/checkTheseTickers.csv")
```

```python
masterTickers.head()
```

```python
checkTheseTickers.head()
```

```python
#apply fuzzy wuzzy string matching 
name = []
similarity = []

for i in masterTickers.symbol_name: 
    ratio = process.extract(i, checkTheseTickers.symbol_name, limit=1) 
    name.append(ratio[0][0])
    similarity.append(ratio[0][1]) 
    
masterTickers['newId'] = pd.Series(name) 
masterTickers['similarity'] = pd.Series(similarity)
masterTickers.head()
```

```python
#export as csv
masterTickers.to_csv(r'Data/Transformed/masterTickers.csv',index=False)
```
